```python
#17 Função com desconto

def calcula_desconto(valor, desconto):
    valor_final = valor - (valor * (desconto / 100))
    return valor_final

# Teste
resultado = calcula_desconto(100, 10)
print(resultado)
```

```python
#18 Classificação de nota


def classifica_nota(nota):
    if nota >= 7:
        return "Aprovado"
    elif nota >= 5:
        return "Recuperação"
    else:
        return "Reprovado"

# Teste
print(classifica_nota(6.5))
```

```python
#19 Contagem de positivos, negativos e zeros.

numeros = [3, -1, 0, 5, -7, 0, 2]

positivos = 0
negativos = 0
zeros = 0

for n in numeros:
    if n > 0:
        positivos += 1
    elif n < 0:
        negativos += 1
    else:
        zeros += 1

print(f"Positivos: {positivos}")
print(f"Negativos: {negativos}")
print(f"Zeros: {zeros}")
```
